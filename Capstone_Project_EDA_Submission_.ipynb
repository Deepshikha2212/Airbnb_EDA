#*11. Highest and Lowest price according to location* 

```python
#To get the highest price we are using groupby and sorting the maximun price.
highest_price_location =data.groupby('neighbourhood_group', as_index=False)['price'].max().sort_values(['price'], ascending = False).rename(columns={'price':'Maximum Price'})
highest_price_location #displaying the table 
```

```python
highest_price_location
```

```python
Highest_price=highest_price_location.plot(kind='bar', x='neighbourhood_group', y='Maximum Price')  #Plotting chart to represent the maximum price 
plt.figure('figsize=(10,5)')  
```

```python
#Using groupby and minimum to get the values 
lowest_price_location =data.groupby('neighbourhood_group', as_index=False)['price'].min().sort_values(['price'], ascending = False).rename(columns={'price':'Minimum Price'})
```

```python
Lowest_price=lowest_price_location.plot(kind='bar', x='neighbourhood_group', y='Minimum Price')  #Plotting chart to represent the maximum price 
plt.figure('figsize=(10,5)') #Display the chart 
```

```python
Merge_price= pd.merge(highest_price_location,lowest_price_location, on='neighbourhood_group') #merging the minimun price and maximun price accordinf to the loacation 
Merge_price #displaying the merged price 
```

```python
Price_Variation=Merge_price.copy()
Price_Variation
```

# *12. Maximum and Minimum Price according to Location* 

```python
#price variation plot keeping neighbourhood in x axis and, multi bar of Maximum and Minimum  
Price_Variation.plot(x='neighbourhood_group',y=['Maximum Price','Minimum Price'], kind='bar',figsize=(10,5))
plt.title('Maximum and Minimum Price according to location')
plt.legend(['Maximum Price','Minimum Price'])
plt.ylabel('Price')
```

**We can see from the graph from the values that the range is high for the highest value and the lowest value so we will apply log to the function**  




```python
def log(values): #creating the function of log to get the chart
  new_price_log=np.log2(values) 
  return new_price_log

Price_Variation['Maximum Price']=Price_Variation.apply(lambda x : log(x['Maximum Price']), axis=1)#Using lambda function 
Price_Variation['Minimum Price']=Price_Variation.apply(lambda x : log(x['Minimum Price']), axis=1)

Price_Variation.rename(columns={'neighbourhood_group':'Location'})
```

```python

Price_Variation.plot(x='neighbourhood_group', y=['Maximum Price','Minimum Price'],kind='bar',figsize=(10,5))
plt.legend(['Maximum Price','Minimum Price'])
plt.title('Maximum and Minimum price according to location')
plt.ylabel('Log Price')
plt.xlabel('Locations')
```
